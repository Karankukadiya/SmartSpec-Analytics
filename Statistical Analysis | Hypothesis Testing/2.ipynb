### Step 1: Import Libraries
Load `pandas`/`numpy`, plotting libraries, and `pearsonr`/`ttest_ind` from `scipy.stats` for correlation and hypothesis testing.

```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy.stats import ttest_ind
```

### Step 2: Load Dataset
Read the smartphone dataset from `Dataset.xlsx`.

```python
df = pd.read_excel('Dataset.xlsx')
```

### Step 3: Descriptive Statistics
Summarize `Price` and `Rating` (count, mean, std, quartiles, min/max).

```python
df[["Price","Rating"]].describe()
```

### Step 4: Scatter Plot — Price vs Rating
Visualize the relationship between price and rating for each phone as a scatter plot.

```python
plt.figure(figsize=(8,6))

plt.scatter(
    df["Price"],
    df["Rating"],
    alpha=0.5
)

plt.xlabel("Current Price")
plt.ylabel("Rating")
plt.title("Price vs Rating")

plt.show()
```

### Step 5: Correlation Matrix
Compute the pairwise Pearson correlation between `Price` and `Rating`.

```python
df[[
    "Price",
    "Rating"
]].corr()
```

### Step 6: Pearson Correlation Test
Run a formal Pearson correlation test between `Price` and `Rating`, returning the correlation coefficient and p-value.

```python
corr, p_value = pearsonr(
    df["Price"],
    df["Rating"]
)
```

### Step 7: Print Test Results
Display the correlation coefficient (r) and p-value from the test.

```python
print("Correlation Coefficient (r):", corr)
print("P-value:", p_value)
```

### Step 8: Hypothesis Decision
Compare the p-value to α = 0.05 and report whether to reject the null hypothesis of no linear relationship.

```python
alpha = 0.05

if p_value < alpha:
    print("Reject H₀")
else:
    print("Fail to Reject H₀")
```

### Business Question

Is there a relationship between smartphone price and customer rating?

### Hypotheses

H₀:

There is no significant linear relationship between smartphone price and customer rating.

H₁:

There is a significant linear relationship between smartphone price and customer rating.

### Statistical Test

Pearson Correlation Test

### Results

Metric	Value
- Correlation Coefficient (r)	0.4827
- P-value	< 0.001

### Decision

Since the p-value is less than 0.05, reject the null hypothesis.

### Business Insight

Smartphone price and customer ratings have a statistically significant moderate positive relationship (r = 0.482, p < 0.001). This indicates that higher-priced smartphones tend to receive better customer ratings. However, the relationship is moderate rather than strong, suggesting that factors such as camera quality, processor performance, battery life, and brand reputation also play important roles in customer satisfaction.
