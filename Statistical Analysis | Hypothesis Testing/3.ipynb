### Step 1: Import Libraries
Load `pandas`/`numpy`, plotting libraries, and `pearsonr`/`ttest_ind` from `scipy.stats`.

```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy.stats import ttest_ind
```

### Step 2: Load Dataset
Read the smartphone dataset from `Dataset.xlsx`.

```python
df = pd.read_excel('Dataset.xlsx')
```

### Step 3: Descriptive Statistics
Summarize `Discount` and `Reviews` columns.

```python
df[["Discount", "Reviews"]].describe()
```

### Step 4: Scatter Plot — Discount vs Reviews
Visualize the relationship between discount percentage and number of reviews.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.scatter(
    df["Discount"],
    df["Reviews"],
    alpha=0.5
)

plt.xlabel("Discount (%)")
plt.ylabel("Review Count")
plt.title("Discount vs Review Count")

plt.show()
```

### Step 5: Correlation Matrix
Compute the pairwise Pearson correlation between `Discount` and `Reviews`.

```python
df[["Discount", "Reviews"]].corr()
```

### Step 6: Pearson Correlation Test
Run a Pearson correlation test between `Discount` and `Reviews`, printing the correlation coefficient and p-value.

```python
corr, p_value = pearsonr(
    df["Discount"],
    df["Reviews"]
)

print("Correlation:", corr)
print("P-value:", p_value)
```

### Business Problem
Does a higher discount increase review count?

### Hypotheses
H₀: There is no significant linear relationship between discount percentage and review count.

H₁: There is a significant linear relationship between discount percentage and review count.

### Statistical Test
Pearson Correlation Test

### Results
Metric	Value
Correlation (r)	-0.110
P-value	1.20 × 10⁻¹⁶

### Decision
Since p-value (1.20 × 10⁻¹⁵) < α (0.05), reject the null hypothesis.

### Business Insight
A statistically significant but very weak negative relationship was found between discount percentage and review count (r = -0.110, p < 0.001). Although the relationship is statistically significant due to the large sample size (n = 5,627), its practical impact is minimal. This suggests that increasing discounts alone is unlikely to meaningfully increase customer review counts. Other factors such as brand reputation, product quality, product age, and sales volume are likely to have a much stronger influence on the number of reviews.
