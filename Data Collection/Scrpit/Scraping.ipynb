```python
from selenium import webdriver
from selenium.webdriver.common.by import By
driver = webdriver.Chrome()
driver.get("https://www.google.com")
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import time
import re
```

```python
# Initialize driver
driver = webdriver.Chrome()
driver.get('https://www.flipkart.com/')

wait = WebDriverWait(driver, 20)

# Close login popup if it appears
try:
    close_btn = wait.until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[5]/div/span"))
    )
    close_btn.click()
except:
    pass  # No popup appeared

# Correct Flipkart search box
search = wait.until(
    EC.element_to_be_clickable((By.NAME, "q"))  # Flipkart uses name="q" for search
)
search.click()
search.clear()
search.send_keys("Smartphones")
search.send_keys(Keys.RETURN)
```

```python
products = []
```

```python
for page in range(1, 22):

    url = f"https://www.flipkart.com/search?q=smartphones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.brand%255B%255D%3DMOTOROLA&page={page}"
    driver.get(url)

    time.sleep(3)

    cards = driver.find_elements(By.CSS_SELECTOR, "div.jIjQ8S")
    print(f"Page {page}: Found {len(cards)} cards")
    for card in cards:
        try:
            product_name = card.find_element(
                By.CSS_SELECTOR,
                "a.k7wcnx div.RG5Slk"
            ).text
        except:
            product_name = None

        try:
            Current_price = card.find_element(
                By.CSS_SELECTOR,
                "div.hZ3P6w"
            ).text
        except:
            Current_price = None

        try:
            mrp = card.find_element(By.CSS_SELECTOR, "div.kRYCnD").text
        except NoSuchElementException:
            mrp = None

        try:
            discount = card.find_element(By.CSS_SELECTOR, "div.HQe8jr span").text
        except:
            discount = None

        try:
            rating = card.find_element(By.CSS_SELECTOR, "div.MKiFS6").text
        except:
            rating = None

        try:
            review_text = card.find_element(By.CSS_SELECTOR, "span.PvbNMB").text
        except:
            review_text = None

        try:
            product_url=card.find_element(By.TAG_NAME,"a").get_attribute("href")
        except:
            product_url=None

        try:
            image_url=card.find_element(By.TAG_NAME,"img").get_attribute("src")
        except:
            image_url=None

        specs = card.find_elements(By.CSS_SELECTOR, "li.DTBslk")
        
        spec_data = {
            "RAM": None,
            "Storage": None,
            "Display": None,
            "Camera": None,
            "Battery": None,
            "Processor": None
        }
        
        for spec in specs:
        
            text = spec.text.strip()
        
            # RAM + Storage
            if "RAM" in text and "ROM" in text:
                parts = [x.strip() for x in text.split("|")]
        
                if len(parts) >= 1:
                    spec_data["RAM"] = parts[0]
        
                if len(parts) >= 2:
                    spec_data["Storage"] = parts[1]
        
            # Display
            elif "Display" in text:
                spec_data["Display"] = text
        
            # Camera
            elif "Camera" in text:
                spec_data["Camera"] = text
        
            # Battery
            elif "mAh" in text:
                spec_data["Battery"] = text
        
            # Processor
            elif ("Processor" in text or
                  "Snapdragon" in text or
                  "Dimensity" in text or
                  "Helio" in text or
                  "Exynos" in text or
                  "Tensor" in text or
                  re.search(r"T\d{4}", text)):
                spec_data["Processor"] = text
                
        products.append({
    
            "Product Name": product_name,
            "Current Price": Current_price,
            "MRP": mrp,
            "Discount": discount,
            "Rating": rating,
            "Review text": review_text,
            
            "RAM": spec_data["RAM"],
            "Storage": spec_data["Storage"],
            "Display": spec_data["Display"],
            "Camera": spec_data["Camera"],
            "Battery": spec_data["Battery"],
            "Processor": spec_data["Processor"],
            
            "Product URL": product_url,
            "Image URL": image_url
            
        })

df = pd.DataFrame(products)
df.head()
df.to_excel("motorola.xlsx", index=False) 
```
