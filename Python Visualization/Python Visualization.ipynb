# Python Visualization

**Import libraries & set plot theme**
Loads `pandas`, `matplotlib`, and `seaborn`, and configures a default seaborn theme (`whitegrid`, `Set2` palette).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Theme
sns.set_theme(style="whitegrid", palette="Set2")

plt.rcParams["figure.figsize"] = (10,6)
plt.rcParams["figure.dpi"] = 120
```

**Load the dataset**
Reads the smartphone dataset from `Dataset.xlsx` into the DataFrame `df`.

```python
df = pd.read_excel('Dataset.xlsx')
```

**Preview the data**
Shows the first 5 rows of `df` to inspect its columns and values.

```python
df.head()
```

**Price distribution**
Histogram (with KDE) of the `Price` column.

## Price Analysis

```python
sns.histplot(df["Price"], bins=30, kde=True)
plt.title("Price Distribution")
plt.show()
```

**Price by brand**
Boxplot showing the spread of `Price` for each `Brand`.

```python
sns.boxplot(data=df, x="Brand", y="Price")
plt.xticks(rotation=45)
plt.title("Price Distribution by Brand")
plt.show()
```

**Average price by brand**
Groups by `Brand`, computes mean `Price`, sorts ascending, and plots as a bar chart.

```python
avg = df.groupby("Brand")["Price"].mean().sort_values()

avg.plot(kind="bar")
plt.title("Average Price by Brand")
plt.ylabel("Price")
plt.show()
```

**Top 10 most expensive smartphones**
Selects the 10 rows with the highest `Price` and plots them as a horizontal bar chart (most expensive at top).

```python
top = df.nlargest(10, "Price")

plt.barh(top["Name"], top["Price"])
plt.title("Top 10 Most Expensive Smartphones")
plt.gca().invert_yaxis()
plt.show()
```

**Top 10 cheapest smartphones**
Selects the 10 rows with the lowest `Price` and plots them as a horizontal bar chart (cheapest at top).

```python
cheap = df.nsmallest(10, "Price")

plt.barh(cheap["Name"], cheap["Price"])
plt.title("Top 10 Cheapest Smartphones")
plt.gca().invert_yaxis()
plt.show()
```

**Correlation heatmap**
Computes the pairwise correlation matrix for `Price`, `RAM`, `Storage`, `Battery`, `Rating`, `Discount`, `Ratings`, and `Reviews`, and visualizes it as an annotated heatmap.

```python
cols = [
    "Price",
    "RAM",
    "Storage",
    "Battery",
    "Rating",
    "Discount",
    "Ratings",
    "Reviews"
]

sns.heatmap(df[cols].corr(),
            annot=True,
            cmap="coolwarm")

plt.show()
```

**Pairplot of key numeric features**
Generates pairwise scatterplots (and diagonal distributions) for `Price`, `RAM`, `Storage`, `Battery`, `Rating`, and `Discount` to spot relationships across multiple variables at once.

```python
sns.pairplot(
    df[[
        "Price",
        "RAM",
        "Storage",
        "Battery",
        "Rating",
        "Discount"
    ]]
)

plt.show()
```

## Performance & Value Metrics

**Value-for-money score by brand**
Creates a new `Value_Score` column (`Rating * 100 / Price`) and plots the average score per `Brand`, sorted ascending — higher bars indicate better rating-per-rupee.

```python
df["Value_Score"] = (df["Rating"] * 100) / df["Price"]

df.groupby("Brand")["Value_Score"].mean().sort_values().plot(kind="bar")
plt.title("Value-for-Money Score")
plt.show()
```

**Battery-to-price ratio by brand**
Creates a new `Battery_Price` column (`Battery / Price`) and plots the average ratio per `Brand`.

```python
df["Battery_Price"] = df["Battery"] / df["Price"]

df.groupby("Brand")["Battery_Price"].mean().plot(kind="bar")
plt.show()
```

**Price vs Rating, colored by brand and sized by RAM**
Scatterplot of `Price` against `Rating`, using `hue` to distinguish `Brand` and point `size` to encode `RAM`.

```python
sns.scatterplot(
    data=df,
    x="Price",
    y="Rating",
    hue="Brand",
    size="RAM"
)

plt.show()
```

**Price density plot**
Filled KDE (density) plot of the `Price` column.

```python
sns.kdeplot(df["Price"], fill=True)
plt.show()
```
